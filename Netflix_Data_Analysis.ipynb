```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading CSV file of dataset
data = pd.read_csv('/content/netflix.csv')
data.head()
```

```python
print('Dataset Shape:', data.shape)
```

```python
#columns and dtypes
data.info()
```

```python
#statistical summary
data.describe(include = 'all')
```

```python
#data preprocessing
```

```python
#Missing values checking
data.isnull().sum()
```

```python
data[data['director'] == 'Not Given']
# Some values in director and country column are given as 'Not Given' which cal also called as null values
```

```python
data[data['country'] == 'Not Given']
```

```python
data['director'] = data['director'].replace('Not Given', np.nan)
data['country'] = data['country'].replace('Not Given', np.nan)
```

```python
data.isnull().sum()
```

```python
# We have null values in director and country column.
#visualize missing values
plt.figure(figsize = (10,6))
sns.heatmap(data.isnull())
plt.title('Missing Values Heatmap')
plt.show()
```

```python
data.dropna(subset = ['director','country'], inplace=True)
```

```python
data.isnull().sum()
```

```python
#checking for duplicates
data.duplicated().sum()
```

```python
# We don't have any duplicates still we use drop duplicates to remove if any
data.drop_duplicates()
```

```python
# Converting data_added column to datetime (dtype)
data['date_added'] = pd.to_datetime(data['date_added'])
```

```python
print(data.dtypes)
```

```python
#for preprocessed data
data.to_csv('netflix1.csv', index = False)
```
